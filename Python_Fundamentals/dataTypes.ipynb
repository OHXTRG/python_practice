## Strings

- Strings are immutable data type . 

```python
my_str = 'greetings'
print(my_str,type(my_str)) 
print('h' in my_str) # in operator
# my_str[1] = 'N' # error can't assing string char
# print(my_str)
my_int = 50 
my_str2 = 'new str'
concat = my_str + my_str2 + str(my_int)
print(concat)

# F strings 
num1 = 50 
num2 = 4
newString = f'The sum {num1} and {num2} will be {num1+num2}'
print(newString)
```

```python
## Accessing strings 
my_str = 'This is world'
print(my_str[0])
print(my_str[-1])
print(my_str[1:4])
print(len(my_str))
print(my_str[-1:-len(my_str)-1:-1]) # string[start:stop:step] stop is exclusive
```

```python
# built in method for strings
my_str = "  hello world   "
my_str_upper = my_str.upper()
print(my_str_upper) 

my_str_lower = my_str.lower()
print(my_str_lower)

my_str_strip = my_str.strip()
print(my_str_strip)

my_str_replace = my_str.replace("l","R")
print(my_str_replace)

my_str_split = my_str.split(" ")
print(my_str_split,type(my_str_split))

my_list = ["hello" , "world" , "Hi" ]
my_str_join = ' '.join(my_list)
print(my_str_join)

start_with_hello = my_str.startswith("hello")
print(start_with_hello)

ends_with_world = my_str.endswith("world   ")
print(ends_with_world)

my_str_find = my_str.find('hello')
print(my_str_find)


my_var = 4 
++my_var
print(my_var)
my_var += 1
print(my_var) 
print(my_var)
```
