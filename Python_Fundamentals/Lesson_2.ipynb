```python
## global and nonlocal keyword
## global makes the variable accessible to all scopes 
## nonlocal  make the enclosed variable accessible to its outer scope


def my_fun():
    global my_var
    my_var = "inside function"
    print(my_var)

my_fun()
my_var = "outside function"
print(my_var)


def outter_func():
    res = "hello"
    print(res)
    def inner_func():
        nonlocal res
        res = "how are you"
    inner_func()
    print(res)



# outter_func() only nonlocal on step outer 

def outter_outter():
    res = "new"
    print(res)
    def outter_func():
        # nonlocal res
        res = "hello"
        print(res)
        def inner_func():
            nonlocal res
            res = "how are you"
        inner_func()
        print(res)
    outter_func()
    print(res)

outter_outter()
```

```python
# apply discount function


def apply_discount(price , discount):
    if(not isinstance(price,int) and not isinstance(price,float)):
        return "The price should be a number"
    
    if(not isinstance(discount,int) and not isinstance(discount,float)):
        return "The discount should be a number"
    
    if(price <= 0):
        return "The price should be greater than 0"
    
    if(discount < 0 or discount > 100):
        return "The discount should be between 0 and 100"
    
    return price - (discount/100)*price


print(apply_discount(74.5, 20.0))

print(apply_discount(50, 0))   
print(apply_discount(200, 50))
print(apply_discount(100, 20))


    
```

```python
# caeser ciper function 

def caesar(text, shift, encrypt=True):

    if not isinstance(shift, int):
        return 'Shift must be an integer value.'

    if shift < 1 or shift > 25:
        return 'Shift must be an integer between 1 and 25.'

    alphabet = 'abcdefghijklmnopqrstuvwxyz'

    if not encrypt:
        shift = - shift
    
    shifted_alphabet = alphabet[shift:] + alphabet[:shift]
    translation_table = str.maketrans(alphabet + alphabet.upper(), shifted_alphabet + shifted_alphabet.upper())
    encrypted_text = text.translate(translation_table)
    return encrypted_text

def encrypt(text, shift):
    return caesar(text, shift)
    
def decrypt(text, shift):
    return caesar(text, shift, encrypt=False)

encrypted_text = 'Pbhentr vf sbhaq va hayvxryl cynprf.'
decrypted_text = decrypt(encrypted_text,13)
print(decrypted_text)
```

```python
## create character function

full_dot = '●●●●●●●●●●'
empty_dot = '○○○○○○○○○○○'




def create_character(name , strength , intelligence , charisma):
    # name = args[0]
    # strength = args[1]
    # intelligence = args[2]
    # charisma = args[3]
    # print(name)

    if not isinstance(name,str): 
        return "The character name should be a string"
    if name == "":
        return "The character should have a name"
    if len(name) > 10:
        return "The character name is too long"
    if name.find(" ") != -1:
        return "The character name should not contain spaces"

    if not isinstance(strength , int) or not isinstance(intelligence,int) or not isinstance(charisma,int):
        return "All stats should be integers"
    if strength < 1 or intelligence < 1 or charisma < 1 : 
        return "All stats should be no less than 1"
    if strength > 4 or intelligence > 4 or charisma > 4 :
        return "All stats should be no more than 4"
    if strength + intelligence + charisma != 7:
        return "The character should start with 7 points"
    charName = ''
    charName += name + "\n"

    charName += 'STR '
    charName += full_dot[0:strength] + empty_dot[strength+1:]


    charName += "\nINT "
    charName += full_dot[0:intelligence] + empty_dot[intelligence+1:]
   

    charName += "\nCHA "
    charName += full_dot[0:charisma] + empty_dot[charisma+1:]

    return charName  
        

charN = create_character('jk',4,2,1)
    
print(charN)

```
